```python
%sql
SHOW TABLES IN main.default;
```

```python
%sql
DESCRIBE TABLE main.default.bronze_sales_transactions;
```

```python
%sql
-- ingestion metadata
SELECT
  COUNT(*)                         AS total_records,
  COUNT(ingestion_timestamp)       AS records_with_ingestion_ts,
  COUNT(source_system)             AS records_with_source
FROM main.default.bronze_sales_transactions;
```

```python
%sql
SELECT
  ingestion_date,
  COUNT(*) AS record_count
FROM main.default.bronze_sales_transactions
GROUP BY ingestion_date
ORDER BY ingestion_date DESC;
```

```python
%sql
SELECT
  (SELECT COUNT(*) FROM main.default.bronze_sales_transactions) AS bronze_count,
  (SELECT COUNT(*) FROM main.default.silver_sales_transactions) AS silver_count;
```

```python
%sql
SELECT *
FROM main.default.gold_daily_sales_summary
ORDER BY sales_date DESC
LIMIT 10;
```

```python
%sql
SELECT
  SUM(total_amount) AS silver_revenue,
  (SELECT SUM(total_revenue) FROM main.default.gold_daily_sales_summary) AS gold_revenue
FROM main.default.silver_sales_transactions;
```

```python
%sql
SELECT
  year_month,
  region,
  country,
  monthly_revenue,
  transaction_count
FROM main.default.gold_monthly_revenue_by_region
ORDER BY year_month DESC;
```

```python
%sql
SELECT
  product_id,
  total_units_sold,
  total_revenue,
  avg_selling_price
FROM main.default.gold_product_performance
ORDER BY total_revenue DESC
LIMIT 10;
```
