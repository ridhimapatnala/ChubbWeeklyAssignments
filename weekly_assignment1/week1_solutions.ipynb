```python
# filter by date ranges
# convert to datetime dtype
df["date"] = pd.to_datetime(df["date"], format="mixed", dayfirst=True)
start_date = "2025-10-05"
end_date = "2025-10-07"
df_date_range = df[
    (df["date"] >= start_date) &
    (df["date"] <= end_date)
]
df_date_range
```

```python
# filter by year
df_2025 = df[df['date'].dt.year == 2025]
df_2025.head()
```

```python
# filter october sales (by month)
df_oct = df[df['date'].dt.month == 10] 
df_oct.head()
```

```python
# filter high value transactions
df_high_value = df[df["total"] > 20]
df_high_value
```

```python
#filter low value transactions
df_low_value = df[df["total"]<5]
df_low_value
```

3. Compute descriptive statistics and group summaries (totals by store, average basket).

```python
df.describe() # basic stats for all numeric columns
```

```python
# average basket
print("Min total:", df["total"].min())
print("Max total:", df["total"].max())
print("Avg total:", df["total"].mean())
```

```python
print("Min unit price:", df["unit_price"].min())
print("Max unit price:", df["unit_price"].max())
print("Avg unit price:", df["unit_price"].mean())
```

```python
# total sales by store
sales_by_store = df.groupby("store_name")["total"].sum()
sales_by_store
```

```python
summary_store = df.groupby(["store_id", "store_name"]).agg(
    total_sales=("total", "sum"),
    num_transactions=("transaction_id", "count"),
    avg_sales=("total", "mean")
)
print(summary_store)
```

```python
total_sales_by_product = df.groupby("product_name")["total"].sum()
print(total_sales_by_product)
```

4. Perform simple cleaning (handle missing values, fix data types, drop duplicates).

```python
# checking for duplicate values
df.isnull().sum() 
```

```python
import numpy as np
# filling in missing promo_code
df["promo_code"] = df["promo_code"].fillna("NO_PROMO")

# filling in missing customer_id and convert to int64
df["customer_id"] = df["customer_id"].replace("", np.nan)
df["customer_id"] = df["customer_id"].fillna(0).astype("Int64")

# convert to datetime dtype
df["date"] = pd.to_datetime(df["date"])

print(df.dtypes)
print(df.isnull().sum())
```

```python
dupes = df.duplicated().sum()
print("No. of duplicated records: ",dupes)

# df.drop_duplicates(inplace=True)
```
